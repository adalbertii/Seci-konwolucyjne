```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
```

```python
#input_image = plt.imread('/content/drive/MyDrive/dane/wmi_png.png')
input_image = plt.imread('/content/drive/MyDrive/dane/tymek-01.jpg')
```

```python
print(f'Input shape: {input_image.shape}')
```

```python
gray_image = rgb2gray(input_image)
small_image = gray_image[::2,::2]
```

```python
from scipy.signal import convolve2d

def apply_kernel_to_image(img, kernel, title=''):

    feature = convolve2d(img, kernel, boundary='symm', mode='same')

    # Plot
    fig = plt.figure(figsize=(20, 10))
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.imshow(img, 'gray')
    ax1.set_title('Obraz zrodlowy', fontsize=15)
    ax1.set_xticks([])
    ax1.set_yticks([])

    ax2 = fig.add_subplot(1, 2, 2)
    ax2.imshow(feature, 'gray')
    ax2.set_title(f'Mapa cech - {title}', fontsize=15)
    ax2.set_xticks([])
    ax2.set_yticks([])
    plt.show()
```

```python
#outline
kernel = np.array([
    [-2, -1, -1],
    [-2, 2, -1],
    [-2, 0.1, -1]])

apply_kernel_to_image(gray_image, kernel, 'filtr 1')
```

```python
#left sobel
kernel = np.array([
    [1, 0, -2],
    [2, 0, -2],
    [1, 0, -1]])

apply_kernel_to_image(small_image, kernel, 'filtr 2')
```

```python
#negative
kernel = np.array([
    [-1, -1, -1],
    [-1, -1, -1],
    [-1, -1, -1]])

apply_kernel_to_image(small_image, kernel, 'filtr 3')
```
